```python
#Its a 3x2 matrix which has 3 stocks and 2 features : Price & Volatility
A = np.array([[10, 0.2], [15, 0.4], [20, 0.6]])

#now a 2x1 matrix of weight vector ( weights for Price & Volatility )
B = np.array([[0.7], [0.3]])
```

```python
#time for matrix multiplication now... what makes it really interesting is with just 1 symbol( @) we do it

scores = A@B     #'@' is the modern Python operator for matrix mult
```

```python
print(f"Stock Scores:\n", scores)
```

```python
#Another highly valued skill is : Projection. Which is to to remove the noise from the data. 
#before we do that, assume Vector B has True Signal and Vector A has Market Noise. We need to find, how much noise is in the signal.
#that's projecting A onto B. ( lets do it man !! ) but wait !
#you need a formula to do that, which is easy : proj = (a*b/||b||^2 )*b
```

```python
def project_vector(a,b):
    #1. dot product of a &b
    dot_ab= np.dot(a,b) 
    
    #2. Square of b( which is basically dot of bxb)
    b_sq = np.dot(b,b)

    #3. scaler projection ( as we still got 1 b left outside the bracket)
    return (dot_ab/b_sq)*b 

    
#now I'll project a random signal not the trend line to test it
signal = np.array([3,4])
trend = np.array([1,0]) #X axis

res = project_vector(signal,trend)
print(f"Projected Vector: {res}")
```
